## Difference between the np.linspace and np.random.uniform

* np.linspace creates the value in the range with the increasing the step size 
* np.random.uniform

```python
i = np.linspace(-np.pi, np.pi,1500)
o = np.random.uniform(low = -np.pi,high = np.pi, size=(1500,))
```

```python
plt.plot(i,marker='o')
```
